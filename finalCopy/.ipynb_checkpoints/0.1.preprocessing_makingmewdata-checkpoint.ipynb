## 1. GDP 

```python
GDP = pd.read_csv(path+'국가통계포털/stat_273601_1.csv')
GDP = GDP.T[1:]
GDP = GDP.reset_index()
GDP.columns = ['year','GDP', '성장율'] 
GDP['GDP'] = GDP['GDP'].str.replace(',','').astype(float)
GDP['성장율'] = GDP['성장율'].astype(float)
GDP['year'] = GDP['year'].astype(int)
```

## 2. 인구 조사 가져와서 총계만 가져오고 reset하고 컬럼명 만들고 year를 int로 바꾸어 2034년 이전만 뺌. 
## 인구는 2034년까지 있어 2034년까지 생성됨. 

```python
people = pd.read_csv(path+'인구/성_및_연령별_추계인구_1세별__5세별____전국_20221010195331.csv', encoding='CP949')
people = people.iloc[:1,3:].T.reset_index()
people.columns=['year','people']
people['year'] = people['year'].astype(int)
people['people'] = people['people'].astype(int)
```

```python
GP= pd.merge(left=people, right=GDP , on='year', how='outer') 
# GDP 미래 채우기 전력수급계획 21년 3.6, 25년 2.4, 28년 2.1, 31년 1.7 34년 1.4 
GP = GP.sort_values(by='year', ascending=True).reset_index(drop=True) 
GP.loc[GP['year']==2021, "성장율"] = 3.6
GP.loc[GP['year']==2025, "성장율"] = 2.4
GP.loc[GP['year']==2028, "성장율"] = 2.1
GP.loc[GP['year']==2031, "성장율"] = 1.7
GP.loc[GP['year']==2034, "성장율"] = 1.4 
GP['성장율'] = GP['성장율'].fillna(method='bfill')
GP['누적성장율'] = GP['성장율'].cumsum()

# GDP 0.8을 GDP동일하게 2020년도까지 채우기  
GP['GDP0.8'] = GP['GDP']
```

```python
for i in range(26,40,1):
    GP.loc[i, 'GDP'] = GP.loc[i-1, 'GDP'] * 1.02  # 2% 성장 가정 
    GP.loc[i, 'GDP0.8'] = GP.loc[i-1, 'GDP0.8'] * 1.008   # 마지막 년도 값 유지 0.8% 
```

```python
# 명목 성장율을 구하고 누적도 구하기 
for i in range(1,40,1):
    GP.loc[i, '명GDP성장율'] = round((GP.loc[i, 'GDP'] - GP.loc[i-1, 'GDP'])/GP.loc[i, 'GDP']*100,1)
GP['명누적성장율'] = GP['명GDP성장율'].cumsum()    

for i in range(1,40,1):
    GP.loc[i, '명GDP0.8성장율'] = round((GP.loc[i, 'GDP0.8'] - GP.loc[i-1, 'GDP0.8'])/GP.loc[i, 'GDP0.8']*100,1)
GP['명0.8누적성장율'] = GP['명GDP0.8성장율'].cumsum()

GP = GP[1:].reset_index(drop=True) 
```

## 3.온도 2020년까지는 월별 기상청 데이터 사용 #2021년부터는 기후변화시나리오 2.6 가져와 채우기 

```python
temp = pd.read_csv(path+'temp/OBS_ASOS_MNH_20221002201454.csv', encoding='CP949')
temp = temp[['일시','평균기온(°C)','평균최고기온(°C)', '평균최저기온(°C)']] #,'최저기온(°C)']] 
temp.columns=['date', 'temp_a_rcp6', 'temp_h_rcp6','temp_l_rcp6'] # , 'temp_lowest'] 
temp['date'] = temp['date'] +'-01'
temp['date'] = pd.to_datetime(temp['date'])
temp = temp[temp['date']< '2021-01-01'].reset_index(drop=True)
```

```python
temp_future1 = pd.read_csv(path+'temp/rcp6.0_전국/TA_DIST_ALL_RCP60_OPTYP_MONTH_2021_2034_MEAN.csv') 
temp_future2 = pd.read_csv(path+'temp/rcp6.0_전국/TAMAX_DIST_ALL_RCP60_OPTYP_MONTH_2021_2034_MEAN.csv') 
temp_future3 = pd.read_csv(path+'temp/rcp6.0_전국/TAMIN_DIST_ALL_RCP60_OPTYP_MONTH_2021_2034_MEAN.csv')
temp_future = pd.merge(left=temp_future1, right=temp_future2, on = "요소") 
temp_future =  pd.merge(left=temp_future, right=temp_future3, on = "요소") 
temp_future = temp_future[1:].reset_index(drop=True) 
temp_future['요소'] = temp_future['요소'].str.replace('년','-').str.replace('월','-01')
temp_future.columns=['date','temp_a_rcp6', 'temp_h_rcp6','temp_l_rcp6'] 
temp_future['date'] = pd.to_datetime(temp_future['date'])
temp_all = pd.concat([temp, temp_future],axis=0)
```

```python
# temp_all과 df_all 합치기 
merged_df = pd.merge(left=df_all, right=temp_all, on='date', how='right') 
merged_df['year'] = merged_df['date'].dt.year
merged_df['month'] = merged_df['date'].dt.month
merged_df_r01 = pd.merge(left= merged_df, right= GP, on='year', how='left')
```

```python
# QVA 를 QVA의 비율 변화율의 평균을 곱해서 채우기 0.004  
for i in range(300,468,1):
    merged_df_r01.loc[i, 'QVA'] = merged_df_r01.loc[i-1, 'QVA'] + (merged_df_r01.loc[i-1, 'QVA']* df_all['QVA'].pct_change().mean()/2) # 부가가치증가율이 작아질것으로 예상하고 1/2로  
for i in range(300,468,1):
    merged_df_r01.loc[i, 'QVA_origin'] = merged_df_r01.loc[i-1, 'QVA'] + (merged_df_r01.loc[i-1, 'QVA']* df_all['QVA'].pct_change().mean())    
```

```python
df_train = merged_df.loc[:299]
preds = ["GAS_PRICE","OIL_PRICE","RP"] 
fh = list(range(1,169))
for pred in preds: 
    y= df_train[pred]
    forecaster = TransformedTargetForecaster(
        [
            ("deseasonalize", Deseasonalizer(model="multiplicative", sp=60)),
            ("detrend", Detrender(forecaster=PolynomialTrendForecaster(degree=1))),
            (
                "forecast",
                make_reduction(
                    GradientBoostingRegressor(),
                    scitype="tabular-regressor",
                    window_length=15,
                    strategy="recursive",
                ),
            ),
        ]
    )

    forecaster.fit(y)
    y_pred = forecaster.predict(fh)
    
    # 모델 저장
    joblib.dump(forecaster, f'/home/kogas/Desktop/finalCopy/2_second_model/model_{pred}.pkl')
    model = joblib.load(f'/home/kogas/Desktop/finalCopy/2_second_model/model_{pred}.pkl')
    y_pred = model.predict(fh)
    plot_series(y,y_pred, labels=["y_train",  "y_pred"])
    print("mean:",y_pred.mean(), "min:",y_pred.min(), "max:",y_pred.max())
    if pred == "GAS_PRICE": 
        merged_df.loc[300:, 'GAS_PRICE'] = y_pred.values 
    elif pred == "OIL_PRICE":
        merged_df.loc[300:, 'OIL_PRICE'] = y_pred.values
    elif pred == "RP":
        merged_df.loc[300:, "RP"] = y_pred.values


# mean: 106.03442869829024 min: 71.91561904358353 max: 127.0888333114406 kneighbor 
# mean: 81.7134112928188 min: 64.37698220397262 max: 100.7361974176191 gradient boosting 
```
